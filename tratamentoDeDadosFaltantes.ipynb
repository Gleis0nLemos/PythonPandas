## Tratamento de dados faltantes

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
dados.head(10)
```

```python
dados.isnull()
```

```python
dados.notnull()
```

```python
dados.info()
```

```python
dados['Valor'].isnull()
```

```python
dados[dados['Valor'].isnull()]
```

```python
A = dados.shape[0]
dados.dropna(subset = ['Valor'], inplace=True)
B = dados.shape[0]
A - B
```

```python
#continuação
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
selecao = (dados['Tipo']=='Apartamento') & (dados['Condominio'].isnull())#essa seleção na real é dos dados que queremos excluir
```

```python
A = dados.shape[0]
dados = dados[~selecao] #inverte os valores booleanos: True > False, False > True
B = dados.shape[0]
A - B
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
#pode se escrever assim: dados.fillna(0, inplace = True)
dados = dados.fillna({'Condominio': 0, 'IPTU': 0})
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
dados[dados['IPTU'].isnull()].shape[0]
```

```python
dados.info()
```

```python
dados.to_csv('dados/aluguel_residencial.csv', sep = ';', index = False)
```
